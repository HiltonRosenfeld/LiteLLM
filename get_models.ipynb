```python
import requests

def get_models(self):
    """ GET list of models from LiteLLM API """
    
    model_list = []
    litellm_api_key = self.api_key
    litellm_api_base = self.litellm_api_base

    url = f"{litellm_api_base}/v1/models"

    try:
        response = requests.get(url)
    
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            data = response.json()
            model_list = data.get("data", [])
            model_ids = [model["id"] for model in model_list]
            return model_ids
        else:
            return(f"Failed to retrieve page. Status code: {response.status_code}")
    
    except requests.exceptions.RequestException as e:
        return(f"An error occurred: {e}")
```

```python
data = type('Data', (object,), {})()
data.api_key = "your_api_key_here"
data.litellm_api_base = "http://0.0.0.0:4000"

get_models(data)
```
